# Running `sat_access.py` and `sat_access.R`

This notebook shows users how to run the command-line scripts [`sat_access.py`](sat_access.py) and [`sat_access.R`](sat_access.R).

It focuses on:

- Checking that Python or R is available
- Installing required packages
- Understanding the command-line arguments
- Running common download and plotting commands
- Platform-specific notes for Linux, macOS, and Windows

## Requirements

### Python

- Python 3.x
- `netCDF4`
- `xarray`
- `matplotlib`
- `cartopy`

### R

- R 4.x
- `argparse`
- `ncdf4`
- `curl`
- `lubridate`
- `reshape2`
- `ggplot2`

## Check your environment

### Python

```bash
python --version
```

If you use Conda, activate your environment first:

```bash
conda activate your_env_name
```

### R

Linux/macOS:

```bash
R --version
```

Windows PowerShell:

```powershell
Rscript --version
```

## Install the required packages

### Python

```bash
pip install xarray matplotlib cartopy netCDF4
```

### R

```bash
Rscript -e "install.packages(c('argparse', 'ncdf4', 'curl', 'ggplot2', 'reshape2'), repos='https://cloud.r-project.org/')"
```

For `sat_access.R` on Linux/macOS, make the script executable:

```bash
chmod +x sat_access.R
```

## Command-line arguments

Both scripts accept these core arguments:

| Argument | Meaning |
| --- | --- |
| `--variable` | Data layer to download, for example `SPM` or `CHLA` |
| `--daterange` | One date or two dates in `YYYY-MM-DD` format |
| `--outputdir` | Local folder where files are written |
| `--overwrite` | Overwrite existing files |
| `--plot` | Plot the downloaded data |
| `--boundingbox` | `lon_min lon_max lat_min lat_max` |

Python-only arguments:

| Argument | Meaning |
| --- | --- |
| `--concatenate` | Concatenate downloaded files into one output |
| `--animate` | Animate the downloaded data |

Short forms also exist in the scripts, but the long-form flags are easier for users to read.

## Python examples

Run these commands from the repository root, or from any folder where `sat_access.py` is available.

### Download a single chlorophyll-a file

```bash
python sat_access.py --variable chla --daterange 2025-10-15 --outputdir .
```

### Download multiple SPM files and plot one

```bash
python sat_access.py --variable spm --daterange 2025-10-15 2025-10-17 --outputdir . --plot True --boundingbox 4 6 42 44
```

### Download multiple SPM files, concatenate them, and animate

```bash
python sat_access.py --variable spm --daterange 2025-10-01 2025-10-31 --outputdir . --concatenate True --animate True --boundingbox 4 6 42 44
```

## R examples

The command-line examples below are for [`sat_access.R`](sat_access.R), not `sat_access_script.R`.

### Linux/macOS: download a single chlorophyll-a file and plot it

```bash
./sat_access.R --variable chla --daterange 2025-09-15 --outputdir . --plot TRUE --boundingbox 4 6 42 44
```

The `./` prefix tells the shell to run the script from the current directory.

### Linux/macOS: download multiple SPM files without plotting

```bash
./sat_access.R --variable spm --daterange 2025-11-15 2025-11-17 --outputdir . --plot FALSE
```

### Windows PowerShell: download a single chlorophyll-a file and plot it

```powershell
Rscript sat_access.R --variable chla --daterange 2025-09-15 --outputdir . --plot TRUE --boundingbox 4 6 42 44
```

## How the workflow maps to the scripts

1. The script reads the CLI arguments.
2. It connects to the FTP source for the requested variable and date range.
3. It downloads the NetCDF file or files to `--outputdir`.
4. If `--plot` is enabled, it plots the requested data.
5. In Python, you can also concatenate outputs and generate an animation.

## Troubleshooting

### Python not found

If `python --version` returns nothing but Python is installed, confirm your PATH or activate your Conda environment:

```bash
conda env list
conda activate your_env_name
```

### `ScipyArrayWrapper` / `oindex` error in Python

Update `xarray` and `netCDF4`:

```bash
pip install --upgrade xarray netCDF4
```

### R not found on Windows

If `Rscript --version` fails, add your R installation `bin` directory to PATH, for example:

```text
C:\Program Files\R\R-4.x.x\bin\
```

Then restart PowerShell.

## Notes

- The first `cartopy` run in Python may download shapefiles, so plotting can look slow at first.
- The Python script displays plots interactively.
- The R script saves plots directly to the output directory.
- If command-line `R` or `Python` are troublesome on your machine, the repository also includes [`sat_access_script.R`](sat_access_script.R) and [`sat_access_script.py`](sat_access_script.py) for IDE-based use.
